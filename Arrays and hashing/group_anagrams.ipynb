```python
# so basically sort the elements first and group them using a hashmap, time-O(m*nlogn) , space-O(m*n)
from collections import defaultdict
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        map=defaultdict(int)
        for i in range(len(nums)):
            if target-nums[i] in map:
                return [map[target-nums[i]],i]
            map[nums[i]]=i    
```

```python
#So say in our case if we dont want to sort we just replace the keys basically with arrays ]
#time-O(m*n)
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        op=defaultdict(list)
        for i in strs:
            countA=[0]*26
            for c in i:
                countA[ord(c)-ord('a')]+=1
            op[tuple(countA)].append(i)    
        return list(op.values())

        
```
