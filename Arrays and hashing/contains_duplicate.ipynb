```python
# used the brute force approach to solve this - time complexity -O(n^2),space complexity -O(1)
class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]==nums[j]:
                    return True
        return False            
```

```python
# so here i just sorted it and solved it - time complexity - O(nlogn) due to sort , space-O(1)
class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        nums.sort()
        for i in range(0,len(nums)-1):
            if nums[i]==nums[i+1]:
                return True
        return False               
```

```python
# Now i used up extra space with the help of a dictionary/hashmap . But was able to solve it in O(n) time 
from collections import defaultdict
class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        map=defaultdict(int)
        for i in range(len(nums)):
            map[nums[i]]=map[nums[i]]+1
        for key,value in map.items():
            if value>=2:
                return True
        return False        
                    
```
