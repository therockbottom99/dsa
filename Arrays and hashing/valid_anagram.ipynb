```python
# used the normal strings methods , used sorting so time -O(n),space-O(1)
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        return ''.join(sorted(s))==''.join(sorted(t))
```

```python
#okay so a hashmap was used here and the time - O(n) and space-O(n)
from collections import defaultdict
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        map1=defaultdict(int)
        map2=defaultdict(int)
        for i in s:
            map1[i]+=1
        for j in t:
            map2[j]+=1
        return map1==map2     
```
