```python
# normal binary search 
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        leftPtr=0
        rightPtr=len(nums)-1
        while leftPtr<=rightPtr:
            mid=(leftPtr+rightPtr)//2
            if nums[mid]==target:
                return mid
            elif nums[mid]>target:
                rightPtr=mid-1
            else:
                leftPtr=mid+1    
        return -1                
```
