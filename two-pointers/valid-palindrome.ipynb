```python
# usage of string methods and two pointers approach -O(n)
#s=''.join(s.split()).lower()
class Solution:
    def isPalindrome(self, s: str) -> bool:
        ns=''
        for i in s:
            if i.isalnum():
                ns+=i.lower()
        leftPtr=0
        rightPtr=len(ns)-1
        while leftPtr<rightPtr:
            if ns[leftPtr]!=ns[rightPtr]:
                return False
            leftPtr+=1
            rightPtr-=1    
        return True                
```
