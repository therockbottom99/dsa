```python
# three sum problem - wuth two pointers - time - O(n^2), space-O(m): for the output list 
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        op=[]
        nums.sort()
        for i in range(len(nums)):
            if i>0 and  nums[i]==nums[i-1]:
                continue

            l,r=i+1,len(nums)-1
            while l<r:
                if nums[l]+nums[r]+nums[i]==0:
                    op.append([nums[i],nums[l],nums[r]])
                    l+=1
                    r-=1
                    if nums[l]==nums[l-1]:
                        l+=1
                    if nums[r]==nums[r+1]:
                        r-=1    
                elif nums[l]+nums[r]+nums[i]>0:
                    r-=1
                elif nums[l]+nums[r]+nums[i]<0:
                    l+=1
        return op



            


```
