```python
# this is the most effienct approach(two pointer-Time-O(n),space-O(1)) , ineffecient approach is using brute force, 
# we can use hashmap as well - which is time-O(n) but space - O(n) and binary search is kinda weird time - O(n*logn) space-O(1)

class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l,r=0,len(numbers)-1
        while l<r:
            if numbers[l]+numbers[r]>target:
                r-=1
            elif numbers[l]+numbers[r]<target:
                l+=1
            elif numbers[l]+numbers[r]==target:
                return [l+1,r+1]    
        return []            
                
```
