```python
# Now this is a brute force solution where we check for every possible option by iterating with two loops
#time - O(n),space-O(1)
#Can we optimise it further?
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        maxValue=0
        for i in range(len(prices)):
            for j in range(i+1,len(prices)):
                maxValue=max(maxValue,prices[j]-prices[i])
        return maxValue        
```

```python
# ok in this solution we just track the min value from each element to the left and subtract it tbh 
# dont know if this follows any specific pattern
# time-O(n),space-O(1)
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        minValue=prices[0]
        maxProfit=0
        for i in range(1,len(prices)):
            minValue=min(minValue,prices[i-1])
            maxProfit=max(maxProfit,prices[i]-minValue)
        return maxProfit    
```
